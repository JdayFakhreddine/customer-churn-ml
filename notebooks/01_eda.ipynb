```python
leakage_cols = [
    "Churn Reason",
    "Churn Score",
    "Churn Value"
]

df_clean = df.drop(columns=leakage_cols)
```

```python
df_clean.isnull().sum().sort_values(ascending=False)
```

```python
df_clean.dtypes
```

```python
num_cols = df_clean.select_dtypes(include=["int64", "float64"]).columns
cat_cols = df_clean.select_dtypes(include=["object"]).columns

num_cols, cat_cols
```

```python
X = df_clean.drop(columns=["Churn Label"])
y = df_clean["Churn Label"].map({"No": 0, "Yes": 1})
```

```python
drop_cols = [
    "CustomerID",
    "Country",
    "State",
    "City",
    "Zip Code",
    "Lat Long"
]

X_ml = X.drop(columns=drop_cols)
```

```python
X_encoded = pd.get_dummies(
    X_ml,
    drop_first=True
)
```

```python
X_encoded.shape
```

```python
X_encoded.head()
```

```python
# Correction du type de Total Charges
df_clean["Total Charges"] = pd.to_numeric(
    df_clean["Total Charges"],
    errors="coerce"
)

# VÃ©rification des NaN crÃ©Ã©s
df_clean["Total Charges"].isnull().sum()
```

```python
df_clean["Total Charges"] = df_clean["Total Charges"].fillna(0)
```

```python
X = df_clean.drop(columns=["Churn Label"])

X_ml = X.drop(columns=drop_cols)

X_encoded = pd.get_dummies(
    X_ml,
    drop_first=True
)
```

```python
X_encoded.shape
```

```python
X_encoded.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)

disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()

plt.show()
```

```python
import pandas as pd

coefficients = pd.DataFrame({
    "Feature": X_train.columns,
    "Importance": model.coef_[0]
})

coefficients = coefficients.sort_values(by="Importance", ascending=False)

coefficients.head(20)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=200, random_state=42)

rf.fit(X_train_scaled, y_train)

y_pred_rf = rf.predict(X_test_scaled)

from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_rf))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=300,
    class_weight="balanced",
    random_state=42
)

rf.fit(X_train_scaled, y_train)

y_pred_rf = rf.predict(X_test_scaled)

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_rf))
```

```python
import numpy as np
from sklearn.metrics import classification_report

# ProbabilitÃ©s churn
y_proba = rf.predict_proba(X_test_scaled)[:,1]

# Nouveau seuil
threshold = 0.3

y_pred_new = (y_proba >= threshold).astype(int)

print(classification_report(y_test, y_pred_new))
```
